# Revenue per user

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick

# Data for plotting
treatment_mean = 3.390
control_mean = 3.374
treatment_EBM = 0.317
control_EBM = 0.325

# x-axis labels and their positions
labels = ['Treatment', 'Control']
x_positions = [0, 1]

# y-axis limits
y_min = 0
y_max = 4

# Figure and axis objects
fig, ax = plt.subplots(figsize=(4, 8))

# Plot the bars with error bars
ax.bar(x_positions, [treatment_mean, control_mean], 
       yerr=[treatment_EBM, control_EBM], capsize=5,
       color=['#76448A', '#B9770E'])

# Set the axis labels and limits
ax.set_xticks(x_positions)
ax.set_xticklabels(labels)
ax.yaxis.set_major_formatter(mtick.StrMethodFormatter('${x:,.2f}'))
ax.set_ylabel('Revenue per User')
ax.set_ylim([y_min, y_max])

# Add a horizontal line at the overall mean
mean = (treatment_mean + control_mean) / 2
ax.axhline(mean, color='gray', linestyle='--')

# Add the value of each bar inside the bar
for i, v in enumerate([treatment_mean, control_mean]):
    ax.text(i, v-1,  '${:.2f}'.format(v), color='white', ha='center', fontweight='bold')

# Add a title
fig.suptitle('Revenue per User: Treatment vs Control', y=0.95, size='x-large')


# Show the plot
plt.show()


# Save the plot as a PNG file
fig.savefig('Revenue_per_User.png', dpi=300, bbox_inches='tight')




```

# Diff. in revenue per user between both groups

```python
# Data for plotting
mean_diff = 0.016
diff_EBM = 0.455


# x-axis labels and their positions
labels = ['Treatment - Control']
x_positions = [0]

# y-axis limits
y_min = -0.5
y_max = 0.5

# Figure and axis objects
fig, ax = plt.subplots(figsize=(4, 8))

# Plot the bars with error bars
ax.bar(x_positions, mean_diff, 
       yerr=diff_EBM, capsize=5,
       color='#21618C')


# Define a custom formatter function
def dollar_formatter(x, pos):
    if x < 0:
        return '-${:,.2f}'.format(abs(x))
    else:
        return '${:,.2f}'.format(x)
    
    
# Set the axis labels and limits
ax.set_xticks(x_positions)
ax.set_xticklabels(labels)
ax.yaxis.set_major_formatter(dollar_formatter)
ax.set_ylabel('Difference in Revenue per User')
ax.set_ylim([y_min, y_max])

# Add a horizontal line at the overall mean
#mean = (treatment_mean + control_mean) / 2
ax.axhline(0, color='#E74C3C', linestyle='--')

# Add the value of each bar inside the bar
ax.text(0, mean_diff+0.02, '${:.2f}'.format(mean_diff), 
        color='grey', va='center_baseline', fontweight='bold', 
        fontdict= {'fontsize':10})

# Add a title
fig.suptitle('Difference in Revenue per User: Treatment - Control', y=0.95, size='x-large')


# Show the plot
plt.show()

# Save the plot as a PNG file
fig.savefig('Diff_revenue_per_User.png', dpi=300, bbox_inches='tight')
```

# CTR

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick

# Data for plotting
treatment_ctr = 4.63
control_ctr = 3.92
treatment_EBM = 0.26
control_EBM = 0.24

# x-axis labels and their positions
labels = ['Treatment', 'Control']
x_positions = [0, 1]

# y-axis limits
y_min = 0
y_max = 5

# Figure and axis objects
fig, ax = plt.subplots(figsize=(4, 8))

# Plot the bars with error bars
ax.bar(x_positions, [treatment_ctr, control_ctr], 
       yerr=[treatment_EBM, control_EBM], capsize=5,
       color=['#76448A', '#B9770E'])

# Set the axis labels and limits
ax.set_xticks(x_positions)
ax.set_xticklabels(labels)
ax.yaxis.set_major_formatter(mtick.StrMethodFormatter('{x:,.1f}%'))
ax.set_ylabel('Click throught rate (CTR)')
ax.set_ylim([y_min, y_max])

# Add a horizontal line at the overall mean
mean = (treatment_ctr + control_ctr) / 2
ax.axhline(mean, color='gray', linestyle='--')

# Add the value of each bar inside the bar
for i, v in enumerate([treatment_ctr, control_ctr]):
    ax.text(x_positions[i], v-2,'{:.2f}%'.format(v), color='white', ha='center', fontweight='bold')

# Add a title
fig.suptitle('CTR: Treatment vs Control', y=0.95, size='x-large') 

# Show the plot
plt.show()


# Save the plot as a PNG file
fig.savefig('CTR.png', dpi=300, bbox_inches='tight')
```

# Diff. in CTR between both groups

```python
# Data for plotting
ctr_diff = 0.71
diff_EBM = 0.358


# x-axis labels and their positions
labels = ['Treatment - Control']
x_positions = [0]

# y-axis limits
y_min = -0.2
y_max = 1.1

# Figure and axis objects
fig, ax = plt.subplots(figsize=(4, 8))

# Plot the bars with error bars
ax.bar(x_positions, ctr_diff, 
       yerr=diff_EBM, capsize=5,
       color='#21618C', width=1)


# Define a custom formatter function
def percent_formatter(x, pos):
    if x < 0:
        return '-{:,.2f}%'.format(abs(x))
    else:
        return '{:,.2f}%'.format(x)
    
    
# Set the axis labels and limits
ax.set_xticks(x_positions)
ax.set_xticklabels(labels)
ax.yaxis.set_major_formatter(percent_formatter)
ax.set_ylabel('Difference in CTR')
ax.set_ylim([y_min, y_max])

# Add a horizontal line at the overall mean
#mean = (treatment_mean + control_mean) / 2
ax.axhline(0, color='#E74C3C', linestyle='--')

# Add the value of each bar inside the bar
ax.text(0, ctr_diff-0.45, '{:.2f}%'.format(ctr_diff), 
        color='white', ha='center', fontweight='bold')

# Add a title
fig.suptitle('Difference in CTR: Treatment - Control', y=0.95, size='x-large')


# Show the plot
plt.show()

# Save the plot as a PNG file
fig.savefig('Diff_CTR.png', dpi=300, bbox_inches='tight')
```
